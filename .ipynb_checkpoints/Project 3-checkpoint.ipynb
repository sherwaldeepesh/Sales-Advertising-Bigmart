```python
#Task5
df_marketing['TotalAdvt'] = sum([df_marketing['TV'],df_marketing['radio'],df_marketing['newspaper']])
```

```python
x = df_marketing['TotalAdvt'].values.reshape(-1,1)
y = df_marketing['sales']

simple_model = linear_model.LinearRegression()

simple_model.fit(x,y)
```

```python
#Intercept
simple_model.intercept_
```

```python
#Coefficient
simple_model.coef_
```

```python
#R2 score
simple_model.score(x,y)
```

```python
#Mathematical Equation to predict Sales 
"0.05083826*x + 3.83"
```

```python
#for advertising value 50000
0.05083826*50000 + 3.83
```

Sales will be 2545.743 lac 

```python
#By the model  predicted value will be
simple_model.predict(50000)
```

```python
#Task 6
xm = df_marketing.iloc[:,1:4].values
ym = df_marketing['sales']

multi_model = linear_model.LinearRegression()

multi_model.fit(xm,ym)
```

```python
#Intercept
multi_model.intercept_
```

```python
#Coefficient
multi_model.coef_
```

```python
#R2 score
multi_model.score(xm,ym)
```

```python
#Mathematical equation to predcit sales
"0.0478062*x1+0.19299749*x2-0.00259761*x3+2.6293"
```

```python
#For Advertising value tv 90000, radio 3000 and newspaper 45000
0.0478062*90000+0.19299749*3000-0.00259761*45000+2.6293
```

4767.287 will be the sales value for the above values

```python
# for Advertising value tv 290000 and newspaper 80000
0.0478062*290000+0.19299749*0-0.00259761*80000+2.6293
```

13658.6185 will be the sales value for above values

```python
#By the model
#For Advertising value tv 90000, radio 3000 and newspaper 45000
multi_model.predict(np.array([90000,3000,45000]).reshape(1,-1))
```

```python
# for Advertising value tv 290000 and newspaper 80000
multi_model.predict(np.array([290000,0,80000]).reshape(1,-1))
```

Both the values are approximate same by mathematical equation as well as model

```python
#Task 7
#for simple_linear_regression_model
X = sm.add_constant(x)
model_ols = sm.OLS(y,X)
result_ols = model_ols.fit()

print(result_ols.summary())
```

```python
#for Multiple_regression_model
X = sm.add_constant(xm)
model_ols = sm.OLS(ym,X)
result_ols = model_ols.fit()

print(result_ols.summary())
```

from the above models summary we can now say that by simple_linear model only we can sure about one variable and the std error is not perfect as the multiple linear model shows that differenciate with multiple values and dependencies
so the multiple model is better than simple linear model

```python
simple_y = simple_model.predict(x)

multi_y = multi_model.predict(xm)
```

```python
simple_mse = metrics.mean_squared_error(y,simple_y)

multi_mse = metrics.mean_squared_error(ym,multi_y)
```

we can see the r2 score of simple linear model is lesser than multiple regression model
that shows directly that the values are also dependent on another attributes also like radio and newspaper and not only independent in pair
as this we see it in pairplot so from these values by comparing with the original we saw that multi_y values are more near to sales in comparison to 
simple_y with sales
Hence we can say that Mutilple Regression model is better choice
